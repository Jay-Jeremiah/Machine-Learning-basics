```python
import numpy
import matplotlib.pyplot as plt
```

```python
rand = numpy.random.uniform(0.0,5.0,250)

print(rand)
```

```python
#visualise in a histogram

plt.hist(rand, 5)
plt.show()
```

```python
#an array with 100000 random numbers between 0.0 and 5.0
rand2 = numpy.random.uniform(0.0,5.0,100000)

#visualing it in a histogram
plt.hist(rand2, 100)
plt.show()
```
